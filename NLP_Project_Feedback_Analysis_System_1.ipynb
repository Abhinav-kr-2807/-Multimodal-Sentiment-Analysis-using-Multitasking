```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
# from nltk.stem import WordNetLemmatizer


# Downloading the  NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Function to perform data preprocessing
def preprocess_text(text):
    # Lowercase the text
    text = text.lower()

    # Removing special characters, numbers, and punctuation
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Tokenizing the text
    tokens = word_tokenize(text)

    # Removing the stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]


    # lemmatizer = WordNetLemmatizer()
    # tokens = [lemmatizer.lemmatize(word) for word in tokens]

    # Joininh the tokens back into a single string
    text = ' '.join(tokens)

    return text

# Applying the preprocessing function to the 'review_description' column
df['review_description'] = df['review_description'].apply(preprocess_text)

# Printing the first few rows of the preprocessed DataFrame
print(df.head())
```

```python
import gensim
from gensim.models import Word2Vec
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize

# Loading our DataFrame here
# df = pd.read_csv("your_dataset.csv")

# Ensuring 'review_description' column contains strings
df['review_description'] = df['review_description'].astype(str)

# Tokenizing the text in the 'review_description' column
df['review_description'] = df['review_description'].apply(word_tokenize)

# Training a Word2Vec model on the tokenized text
# We can adjust the parameters (e.g., vector size, window size) as needed
model = Word2Vec(df['review_description'], vector_size=100, window=5, min_count=1, sg=0)

# Function to obtain the vector representation of a list of tokens
def tokens_to_vector(tokens):
    vector = []
    for token in tokens:
        if token in model.wv:
            vector.append(model.wv[token])
    return vector

# Applying the tokens_to_vector function to create word embeddings for each list of tokens
df['review_embedding'] = df['review_description'].apply(tokens_to_vector)

# Printing the new embedded dataset
print(df[['review_description', 'review_embedding']].head())
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Downloading the VADER lexicon if you haven't already
nltk.download('vader_lexicon')

# Initializing the VADER sentiment intensity analyzer
sia = SentimentIntensityAnalyzer()

# Function to get sentiment scores for a list of tokens
def get_sentiment_scores(tokens):
    # Joining the tokens into a single string
    text = ' '.join(tokens)

    # Calculating sentiment scores using VADER
    sentiment_scores = sia.polarity_scores(text)
    return sentiment_scores

# Applying the function to the 'review_description' column and store the results in a new column
new_df['sentiment_scores'] = new_df['review_description'].apply(get_sentiment_scores)

# Extracting the compound sentiment score (a normalized score representing overall sentiment)
new_df['compound_sentiment'] = new_df['sentiment_scores'].apply(lambda x: x['compound'])

# Printing the new DataFrame with sentiment scores
print(new_df[['review_description', 'compound_sentiment', 'rating']].head())
```

```python
from sklearn.pipeline import FeatureUnion, Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
import numpy as np

# Defining a transformer to convert NumPy arrays to strings
class NumpyArrayToString(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        return [' '.join(map(str, x)) for x in X]

# Defining a functions to select numeric and text features
class ColumnSelector(BaseEstimator, TransformerMixin):
    def __init__(self, columns):
        self.columns = columns

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        return X[self.columns]

numeric_features = ['compound_sentiment', 'neg_score', 'neu_score', 'pos_score']
text_feature = 'review_embedding'
target_column = 'rating'  # Update to 'rating'

# Creating a FeatureUnion pipeline
process_and_join_features = Pipeline([
    ('features', FeatureUnion([
        ('numeric_features', Pipeline([
            ('selector', ColumnSelector(columns=numeric_features))
        ])),
        ('text_features', Pipeline([
            ('selector', ColumnSelector(columns=text_feature)),
            ('to_string', NumpyArrayToString()),  # Converting NumPy arrays to strings
            ('vec', TfidfVectorizer()),  # Useing TfidfVectorizer
        ]))
    ])),
    ('clf', OneVsRestClassifier(RandomForestClassifier(max_depth=None, n_estimators=100, class_weight='balanced')))
])

columns_to_include = ['compound_sentiment', 'neg_score', 'neu_score', 'pos_score', 'review_embedding', 'rating']

# Selecting only the first 1000 rows and specified columns
subset_data = num_data[columns_to_include].iloc[:1000]

# Spliting we data into X (features) and y (target)
X = subset_data.drop(target_column, axis=1)
y = subset_data[target_column]

# Training our machine learning model
process_and_join_features.fit(X, y)

# Performing GridSearchCV with the best parameters
param_grid = {
    'clf__estimator__n_estimators': [100, 200],
    'clf__estimator__max_depth': [None, 10],
}

grid_search = GridSearchCV(process_and_join_features, param_grid, cv=5, verbose=1)
grid_search.fit(X, y)
print("Best Parameters:", grid_search.best_params_)

# Now, evaluating the model on the test set
y_pred_rf = grid_search.predict(X)

# Handling Undefined Metrics
# report_rf = classification_report(y, y_pred_rf, zero_division=1)
# print('Random Forest Classification Report:\n', report_rf)
```
